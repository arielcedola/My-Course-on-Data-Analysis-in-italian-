```python
# Violin plot
fig = plt.figure(figsize=(8,6))
ax = fig.gca()
sns.set_style("whitegrid")
sns.violinplot(x = 'cylinders', y = 'price', data = macchine2, ax = ax)
ax.set_xlabel('Cylinders') # Set text for the x axis
ax.set_ylabel('Price')
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca()
sns.set_style("whitegrid")
sns.violinplot(data = macchine2, x = 'cylinders', y = 'price', hue='aspiration',  split=True, ax = ax)
ax.set_xlabel('Cylinders') # Set text for the x axis
ax.set_ylabel('Price')
```

```python
# Grafico a dispersione, prezzi vs mpg
fig = plt.figure(figsize=(6, 6))
ax = fig.gca()            
macchine2.plot(kind = 'scatter', x = 'citympg', y = 'price', ax = ax)
ax.set_title('Scatter plot di price vs MPG')
ax.set_xlabel('City MPG')
ax.set_ylabel('Auto Price')
```

```python
# Grafico a dispersione, prezzi vs mpg utilizzando aesthetics
def auto_color(df, plot_cols):
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    for col in plot_cols:
        fig = plt.figure(figsize=(10,10))
        ax = fig.gca()
        temp1 = df.loc[df['fueltype'] == 'gas']       
        temp2 = df.loc[df['fueltype'] == 'diesel']
        if temp1.shape[0] > 0:                    
            temp1.plot(kind = 'scatter', x = col, y = 'price' , 
                           ax = ax, color = 'DarkBlue', alpha = 0.3, s=50)                          
        if temp2.shape[0] > 0:                    
            temp2.plot(kind = 'scatter', x = col, y = 'price' , 
                           ax = ax, color = 'Red', alpha = 0.3, s=50) 
        ax.set_title('Grafico a dispersione di price vs. ' + col)
        red_patch = mpatches.Patch(color='Red', label='Diesel')   
        blue_patch = mpatches.Patch(color='DarkBlue', label='Gas')
        plt.legend(handles=[red_patch, blue_patch])
    return plot_cols
auto_color(macchine2, ['citympg'])
```

```python
# Grafico a dispersione, prezzi vs mpg utilizzando aesthetics
def auto_size_2(df, plot_cols):
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    for col in plot_cols:
        fig = plt.figure(figsize=(10,10))
        ax = fig.gca()
        temp1 = df.loc[df['fueltype'] == 'gas']       
        temp2 = df.loc[df['fueltype'] == 'diesel']
        if temp1.shape[0] > 0:                    
            temp1.plot(kind = 'scatter', x = col, y = 'price' , 
                           ax = ax, color = 'DarkBlue', s= .02 * macchine2['enginesize'] * macchine2['enginesize'], 
                       alpha = 0.3)                          
        if temp2.shape[0] > 0:                    
            temp2.plot(kind = 'scatter', x = col, y = 'price' , 
                           ax = ax, color = 'Red', s= .02 * macchine2['enginesize']  * macchine2['enginesize'], 
                       alpha = 0.3) 
    ax.set_title('Grafico a dispersione di price vs. ' + col + ' con engine size')
    red_patch = mpatches.Patch(color='Red', label='Diesel')   
    blue_patch = mpatches.Patch(color='DarkBlue', label='Gas')
    plt.legend(handles=[red_patch, blue_patch])
    return 'Done'
auto_size_2(macchine2, ['citympg']) 
```

```python
# Grafico a dispersione, prezzi vs mpg utilizzando aesthetics
def auto_shape(df):
    import matplotlib.pyplot as plt
    import matplotlib.lines as mlines
    aspiration = ['turbo', 'turbo', 'std', 'std']
    fuel = ['diesel', 'gas', 'diesel', 'gas']
    mrks = ['o', 'o', '+', '+']
    cols = ['Red', 'DarkBlue', 'Red', 'DarkBlue']
    ldg = []
    fig = plt.figure(figsize=(10,10))
    ax = fig.gca()
    for asp, fl, mk, cl in zip(aspiration, fuel, mrks, cols):
        temp = df.loc[(df['aspiration'] == asp) & (df['fueltype'] == fl)]       
        if temp.shape[0] > 0:                    
            temp.plot(kind = 'scatter', x = 'citympg', y = 'price' , ax = ax, color = cl, marker = mk,
                       s= .02 * df['enginesize'] * df['enginesize'], alpha = 0.5)    
        ldg.append(mlines.Line2D([], [], color=cl, marker=mk, markersize=10, 
                                 label=(asp + ' and ' + fl)))    
    ax.set_title('Grafico a dispersione di price vs. city mpg con marker size = engine size')
    plt.legend(handles=ldg)
    return'Done'
auto_shape(macchine2)
```
