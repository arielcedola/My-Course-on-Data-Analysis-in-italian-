```python
# Integrazione dei dati
data = pd.merge(macchine, fabbriche, on = 'make-id', how = 'left')
data
```

```python
# Eliminazione colonne non utilizzate
drop_list = ['symboling', 'normalized-losses', 'make-id']
data.drop(drop_list, axis = 1, inplace = True)
data
```

```python
colonne =  ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base',
       'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke',
       'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
```

```python
# Riordinamento di colonne
data = data[colonne]
data.head()
```

```python
colonne_rename =  ['make', 'fueltype', 'aspiration', 'doors', 'body', 'drive', 'engineloc', 'wheelbase',
       'length', 'width', 'height', 'weight', 'enginetype', 'cylinders', 'enginesize', 'fuelsystem', 'bore', 'stroke',
       'compression', 'horsepower', 'rpm', 'citympg', 'highwaympg', 'price']
```

```python
# Renaming di colonne
data.columns = colonne_rename
data.head()
```

```python
# Quanti dati mancanti ci sono?
data.isnull().sum()
```

```python
# Pulizia di dati mancanti
data = data[~pd.isnull(data).any(axis=1)]
data
```

```python
data.reset_index(drop=True, inplace=True)
data
```

```python
# Eliminazione di righe duplicati
data.drop_duplicates(inplace = True)
```
